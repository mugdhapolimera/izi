```python
from astropy.table import Table
import pandas as pd
import numpy as np
import re

gridfile = 'C:/Users/mugdhapolimera/github/izi/richardson_agnfrac-0-1_binary_csf_n1e2_40.0Myr.csv'
df = pd.read_csv(gridfile)
df = df.dropna()
old_name = ['O\S+3726\S+', 'O\S+3728\S+', 'N\S+386\S+', '\S+4363\S+', 'H\S+4340\S+', 
            'H\S+468\S+', 'A\S+4711\S+', '\S+4861\S+', 'O\S+4958\S+', 'O\S+500\S+', '\S+5875\S+', 'O\S+6300\S+', 
            '\S+6548\S+', 'H\S+656\S+', 'N\S+658\S+', 'S\S+671\S+', 'S\S+673\S+', 'A\S+713\S+']
new_name = ['oii3726', 'oii3729', 'neiii3869', 'oiii4363', 'hgamma', 
        'heii4685', 'ariv4711', 'hbeta', 'oiii4959', 'oiii5007', 'hei5875', 'oi6300', 
        'nii6548', 'halpha', 'nii6584', 'sii6717', 'sii6731', 'ariii7136']

drop_cols = []

for col in df.columns:
    found = 0
    for i in range(len(old_name)):
        if re.findall(old_name[i], col):
            df.rename(columns={col:new_name[i]}, inplace=True)
            found = 1
    if col not in ['AGNFRAC', 'LOGZ', 'LOGQ'] and not(found):
            drop_cols.append(col)
            
df = df.drop(columns = drop_cols)
new_gridfile = r'C:/Users/mugdhapolimera/github/izi/richardson_agnfrac-0-1_binary_csf_n1e2_40.0Myr_lines.fits'
t = Table.from_pandas(df)
print t
print t.keys()
t.write(new_gridfile, format='fits')
```
