```python
from astropy.table import Table
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/mugdhapolimera/github/izi/IZI-richardson_bpass_binary_csf_n1e2_40.0Myr-GrovesDepN-0_109.csv')
t = Table.from_pandas(df)

for i in range(len(t)):
    if i > 1:
        t['FLUX'][i] = float(t['FLUX'][i][3:])
    else:
        t['FLUX'][i] = float(t['FLUX'][i][2:])
    t['Unnamed: 39'] = float(df['Unnamed: 39'][i][:-2])
t['FLUX'] = t['FLUX'].astype(float)

flux = np.column_stack((t['FLUX'],t['Unnamed: 6']))
for i in range(7,40):
    colname = 'Unnamed: '+str(i)
    flux = np.column_stack((flux,t[colname]))
print np.shape(flux)
lines = "cii1335  civ1548  civ1551  ciii1909 ciii1911 cii2324  oii3726  oii3729  neiii3869 sii4069  hgamma   oiii4363 hei4471  hbeta    oiii4959 oiii5007 hei5016  ariii5192 ni5198   nii5755  hei5875  oi6300   siii6312 nii6548  halpha   nii6584  hei6678  sii6717  sii6731  ariii7136 oii7318  oii7320  ariii7751 siii9068 siii9532"
line_id = np.array(lines.split())
idcol = np.repeat(line_id[np.newaxis,:],len(t),0)
col_names = ['NAME','LOGOHSUN','LOGZ','LOGQ','ID','FLUX']
t2 = Table([t['NAME'], t['LOGOHSUN'], t['LOGZ'], t['LOGQ'], idcol,flux], names=(col_names))
t2['FLUX'] = t2['FLUX'].astype(float)
t2
t2.write('IZI-richardson_bpass_binary_csf_n1e2_40.0Myr-GrovesDepN-0_109.fits', format='fits')
```
